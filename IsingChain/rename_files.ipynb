```python
import os
```

```python
for filename in os.listdir('./jobs') :
    print(filename)
    if 'test' in filename : continue
    old_path = os.path.join('./jobs', filename)
    if os.path.isfile(old_path) :
        new_path = os.path.join('./jobs', f'rep_{0}_{filename}')
        os.rename(old_path, new_path)
```

```python
for filename in os.listdir('./data') :
    if 'energy' not in filename : continue
    print(filename)
    old_path = os.path.join('./data', filename)
    if os.path.isfile(old_path) :
        new_path = os.path.join('./data', f'rep_{0}_{filename}')
        os.rename(old_path, new_path)
```
